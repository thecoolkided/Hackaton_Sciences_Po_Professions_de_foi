# Batch Test

## OCR Mistral 

```python
import pandas as pd
```

```python
batch_test = pd.read_csv("/Users/Manon/Desktop/Fac/ENC/Hackaton/SciencesPo/Data_OCR/mistral_ocr_result_clean.csv")
```

```python
batch_test.head()
```

```python
batch_test_0 = batch_test.sample(n=550)
```

```python
batch_test_0
```

```python
for i in range (50):
    batch_test = batch_test_0.iloc[i * 11 : (i + 1) * 11]
    assert(len(batch_test)==11)
    batch_test.to_csv("batch_test_" + str(i), index=False) 
```
