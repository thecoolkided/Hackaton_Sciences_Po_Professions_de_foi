Separer les fichiers _falc de l'ensemble des fichiers PDF 

```python
base_dir = os.path.expanduser("~/Documents/hackaton_week/ScincePo_data")
#dossier avec les fichiers PDF

falc_dir = os.path.join(base_dir, "_falc")
os.makedirs(falc_dir, exist_ok=True)
#pour ne pas perdre les données, on va transferer les fichiers _falc dans un dossier à part
```

```python
total_pdf = 0
falc_pdf = 0
```

```python
for root, dirs, files in os.walk(base_dir):
    # éviter de parcourir le dossier _falc lui-même
    if root.startswith(falc_dir):
        continue

    for file in files:
        if file.lower().endswith(".pdf"):
            total_pdf += 1

            if file.lower().endswith("_falc.pdf"):
                falc_pdf += 1
                src_path = os.path.join(root, file)
                dst_path = os.path.join(falc_dir, file)

                if not os.path.exists(dst_path):
                    shutil.move(src_path, dst_path)
                else:
                    print(f"Fichier déjà existant, ignoré : {file}")

non_falc_pdf = total_pdf - falc_pdf

print("Traitement terminé")
print(f"Nombre total de fichiers PDF : {total_pdf}")
print(f"Nombre de fichiers PDF FALC : {falc_pdf}")
print(f"Nombre de fichiers PDF sans FALC : {non_falc_pdf}")
```
